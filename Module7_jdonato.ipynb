```python
#import packages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#load make_blobs to simulate data
from sklearn.datasets import make_blobs
#load decomp to do PCA analysis with sklearn
from sklearn import decomposition
```

```python
X1, Y1 = make_blobs(n_features=25, 
                    n_samples=5000, 
                    centers=4, 
                    random_state=4, 
                    cluster_std=2)

print(X1.shape)
```

```python
pca = decomposition.PCA(n_components=5)
pc= pca.fit_transform(X1)
pc_df = pd.DataFrame(data = pc, columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
pc_df['Cluster'] = Y1
pc_df.head()
```

```python
print(pca.explained_variance_ratio_)
```

```python
df = pd.DataFrame({'Variance':pca.explained_variance_ratio_, 
                  'PC':['PC1', 'PC2', 'PC3', 'PC4', 'PC5']})
sns.barplot(x='PC', y="Variance", data=df);
```

### It was interesting to find that the more clusters that were added the data did not over lap like it did with less custers
